```python
import pandas as pd
import joblib
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
df = pd.read_csv("data/chennai_houseing_sale.csv")
```

Data Pre-Processing

```python
df = df[['AREA', 'INT_SQFT', 'DATE_SALE', 'N_BEDROOM', 'N_BATHROOM', 'N_ROOM', 'PARK_FACIL', 'DATE_BUILD', 'SALES_PRICE']]
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
df['AREA'].unique()
#Karapakkam - Karapakam
#AnnaNagar - Anna Nagar, Ana Nagar, Ann Nagar
#Adyar - Adyr
#Velachery - Velchery
#Chromepet - Chrompt, Chrmpet, Chormpet, Chrompet
#KKNagar - KK Nagar
#TNagar -T Nagar
```

```python
class_mapping = {
    'Karapakam': 'Karapakkam',
    'Anna Nagar|Ana Nagar|Ann Nagar': 'AnnaNagar',
    'Adyr': 'Adyar',
    'Velchery': 'Velachery',
    'Chrompt|Chrmpet|Chormpet|Chrompet': 'Chromepet',
    'KK Nagar': 'KKNagar',
    'T Nagar': 'TNagar'
}

# Apply the mapping using regex
df['AREA'] = df['AREA'].replace(class_mapping, regex=True)
df['AREA'].unique()
```

```python
df.info()
```

```python
df["DATE_SALE"] = pd.to_datetime(df["DATE_SALE"])
df["DATE_BUILD"] = pd.to_datetime(df["DATE_BUILD"])

df["HOUSE_AGE"] = (df['DATE_SALE'] - df["DATE_BUILD"]).dt.days // 365
df.drop(columns=["DATE_SALE", "DATE_BUILD"], inplace=True)

df.info()
```

```python
df.describe()
```

```python
df.head(10)
```

```python
df['PARK_FACIL'].replace({"Yes":1, "No": 0, "Noo": 0}, inplace=True)
df.head(10)
```

```python
df['N_OTHER_ROOM'] = (df['N_ROOM'] - df['N_BEDROOM'] - df['N_BATHROOM'])
df.drop(columns=['N_ROOM'], inplace=True)
df.head(5)
```

```python
df = df[['AREA', 'INT_SQFT', 'N_BEDROOM', 'N_BATHROOM', 'N_OTHER_ROOM', 'PARK_FACIL', 'HOUSE_AGE', 'SALES_PRICE']]
df.head(5)
```

```python
df[['N_BEDROOM','N_BATHROOM', 'N_OTHER_ROOM']] = df[['N_BEDROOM','N_BATHROOM', 'N_OTHER_ROOM']].astype('int64')
df.head(5)
```

```python
df.to_csv('data/processed_chennai_houseing_sale.csv', index=False)
```

```python
df = pd.get_dummies(df, columns=['AREA'])
df.tail(5)
```

Train Model

```python
y = df['SALES_PRICE']
x = df.drop(columns=['SALES_PRICE'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
house_price_predictor = LinearRegression()
house_price_predictor.fit(x_train, y_train)
```

```python
y_pred = house_price_predictor.predict(x_test).astype(int)
accuracy = r2_score(y_test, y_pred)
accuracy
```

```python
loss = mean_squared_error(y_test, y_pred)
loss
```

```python
house_data = [1000, 1, 1, 3, 1, 15, 0, 0, 1, 0, 0, 0, 0]
house_price = house_price_predictor.predict([house_data])
house_price[0]
```

Save model

```python
file_name = 'houe_predictor_lr.mdl'
joblib.dump(house_price_predictor, file_name)
```
